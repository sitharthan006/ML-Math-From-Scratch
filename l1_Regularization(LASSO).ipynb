# Lasso Regression (L1 Regularization)

## What is Lasso Regression?

Lasso (Least Absolute Shrinkage and Selection Operator) is a regularized version of Linear Regression that introduces an **L1 penalty** into the loss function.

Unlike Ridge Regression, which only shrinks coefficients, Lasso can shrink some coefficients **exactly to zero**, automatically removing unimportant features from the model. For this reason, Lasso performs both **regression** and **feature selection**.

---

## Loss Function

The objective function of Lasso Regression is

$$
J(B)=\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat y_i)^2+\lambda\sum_{j=1}^{p}|B_j|
$$

where

- First term → Mean Squared Error (prediction error)
- Second term → L1 Regularization
- $\lambda$ controls the amount of regularization.

---

## Why L1 Instead of L2?

Unlike Ridge Regression, Lasso uses the **absolute value** of the coefficients instead of their squares.

Since the penalty depends on the absolute value,

$$
|B_j|
$$

there is **no squaring involved**.

Therefore, there is **no power rule**, and consequently **no factor of 2** appears during differentiation.

---

## Derivative of the L1 Penalty

The derivative of the absolute value function is

$$
\frac{d}{dB_j}|B_j|
=
\begin{cases}
+1, & B_j>0\\
-1, & B_j<0\\
\text{undefined}, & B_j=0
\end{cases}
$$

Notice that the derivative is **not defined exactly at zero**, because the absolute value function has a sharp corner.

---

## Piecewise Derivation

Since the absolute value function behaves differently depending on the sign of the coefficient, the loss function must be differentiated separately.

### Case 1 : $B_j>0$

The derivative becomes

$$
\frac{\partial J}{\partial B_j}
=
\frac{\partial MSE}{\partial B_j}
+\lambda
$$

Setting it equal to zero,

$$
\frac{\partial MSE}{\partial B_j}
=
-\lambda
$$

---

### Case 2 : $B_j<0$

The derivative becomes

$$
\frac{\partial J}{\partial B_j}
=
\frac{\partial MSE}{\partial B_j}
-\lambda
$$

Setting it equal to zero,

$$
\frac{\partial MSE}{\partial B_j}
=
\lambda
$$

---

## Why Doesn't Lasso Have a Closed-Form Solution?

Although we can derive the stationary conditions separately for

- $B_j>0$
- $B_j<0$

there is one important problem.

At

$$
B_j=0
$$

the derivative does **not exist**.

A closed-form solution requires the objective function to be differentiable everywhere so that one smooth equation can be solved analytically.

Since Lasso is **not differentiable at zero**, the same analytical approach used for Ridge Regression cannot be applied.

Therefore, Lasso **does not have a closed-form solution**.

Instead, it is solved using iterative optimization algorithms such as **Coordinate Descent** together with **Soft Thresholding**.

---

# Why Can Lasso Set Coefficients Exactly to Zero? (Optimization Intuition)

The gradient of Ridge Regression is

$$
2\lambda B_j
$$

which depends on the current value of the coefficient.

As the coefficient becomes smaller, the regularization gradient also becomes smaller, causing Ridge to gradually stop shrinking before reaching zero.

Lasso behaves differently.

Its regularization gradient is

$$
\lambda \times \operatorname{sign}(B_j)
$$

whose magnitude remains constant regardless of how small the coefficient becomes.

Therefore, the shrinking force does **not weaken** as the coefficient approaches zero.

This allows the optimizer to continue shrinking the coefficient until it reaches

$$
B_j=0
$$

which effectively removes that feature from the model.

---

## Why Does a Zero Coefficient Remove a Feature?

Suppose the prediction equation is

$$
\hat y
=
B_0
+
0\times x_1
+
2x_2
+
5x_3
$$

Since

$$
0\times x_1=0
$$

the value of $x_1$ has **no influence** on the prediction.

Therefore, that feature contributes nothing to the model and is effectively removed.

This is why Lasso is said to perform **automatic feature selection**.

---

## Implementation

Unlike Ridge Regression, Lasso cannot be implemented using a Normal Equation because no analytical solution exists.

In practice, machine learning libraries such as **Scikit-Learn** optimize the Lasso objective using **Coordinate Descent**.

Therefore, this notebook focuses on understanding the mathematics and intuition behind Lasso Regression while using Scikit-Learn for the optimization step.


## Observation

Lasso Regression differs fundamentally from Ridge Regression.

- Ridge shrinks coefficients but usually keeps every feature.
- Lasso can shrink coefficients exactly to zero, automatically removing unimportant features.
- Since the L1 penalty is not differentiable at zero, Lasso is solved using iterative optimization methods instead of a closed-form solution.

## Implementation

Unlike Ridge Regression, Lasso does not have a closed-form analytical solution because the L1 penalty is not differentiable at zero.

In practice, machine learning libraries such as Scikit-Learn solve this optimization problem using **Coordinate Descent**. Therefore, this notebook focuses on understanding the mathematics behind Lasso while using Scikit-Learn's optimized implementation for training.

## Sklearn Implementation

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
x,y=make_regression(n_samples=300,n_features=4,n_informative=2,noise=20,random_state=42)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

lasso = Lasso(alpha=1.0)
lasso.fit(x_train, y_train)
pred = lasso.predict(x_test)

print("Metrics :")
print("MSE",mean_squared_error(y_test,pred))
print("MAE",mean_absolute_error(y_test,pred))
print("R2_SCORE",r2_score(y_test,pred))
```
